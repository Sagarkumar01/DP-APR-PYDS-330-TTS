```python
import cv2
import numpy as py
import matplotlib.pyplot as plt
```

```python
#load and display an image
img = cv2.imread("IMG_20220425_230101.JPG")
plt.imshow(img)
```

```python
rotate_clockwise=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(rotate_clockwise)
```

```python
rotate_clockwise=cv2.rotate(img,cv2.ROTATE_180)
plt.imshow(rotate_clockwise)
```

```python
print(img.shape)
```

```python
#blur the image
blur = cv2.blur(img,(50,70))
plt.imshow(blur)
```

```python
#Draw on image
start_point = (0, 0)
end_point = (250,250)
clr = (0,255,0)
line_thickness = 15
line = cv2.line(img, start_point, end_point, clr, line_thickness)
plt.imshow(line)
```

```python
#Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
#Threshold
ret, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
plt.imshow(thresh)
```

```python
#edgedetection
edges = cv2.Canny(img,100,200)
plt.imshow(edges)
```
